# Create Extra Visualizations

```python
# import dependencies and data
import pandas as pd
import matplotlib.pyplot as plt
import cartopy.crs as ccrs

df = pd.read_csv("../data/cities.csv")
df
```

```python
df['Latn'] = [float(lat) for lat in df['Lat']]
df['Lngn'] = [float(lng) for lng in df['Lng']]
df = df.dropna()
df
```

```python
# create scatter of latitude and longitude
xvals = df['Lng']
yvals = df['Lat']
plt.figure(figsize=(20,10))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.stock_img()
plt.scatter(xvals,yvals,color='red')
plt.title('Latitude and Longitude of Locations Used')
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.ylim(-90,90)
plt.xlim(-180,180)
plt.savefig("latlngmap.png")
plt.show()
```

```python
# create histogram of latitude
plt.figure(figsize=(20,10))
plt.hist(yvals)
plt.title('Histogram of Latitudes for Locations Used')
plt.xlabel('Latitude')
plt.ylabel('Frequency')
plt.xlim(-90,90)
plt.savefig("lathisto.png")
plt.show()
```
